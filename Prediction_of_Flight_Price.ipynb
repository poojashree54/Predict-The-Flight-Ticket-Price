```python
flight['Date_of_Journey']=pd.to_datetime(flight['Date_of_Journey'])
```

```python
def convert_min(duration):
  total_min=0
  if 'h' in duration:
    hours=duration.split('h')[0].strip()
    total_min+=int(hours)*60
  if 'm' in duration:
    minute=duration.split('m')[0].split()[-1].strip()
    total_min+=int(minute)
  return total_min
```

```python
# Assuming your DataFrame is named 'flight' and the duration column is 'Duration'
flight['Duration_minutes'] = flight['Duration'].apply(convert_min)
```

```python
flight['Duration_minutes'].head()
```

```python
flight['month']=flight['Date_of_Journey'].dt.month
```

```python
flight['month'].head()
```

```python
flight['year']=flight['Date_of_Journey'].dt.year
```

```python
flight['Total_Stops']=flight['Total_Stops'].replace('non-stop','0 stop')
```

```python
flight['Total_Stops']=flight['Total_Stops'].str.split(' ').str[0]
```

```python
flight.head()
```

```python
flight.drop(['Route','Arrival_Time','Date_of_Journey'],axis=1,inplace=True)
```

```python
flight.drop(['Dep_Time','Duration'],axis=1,inplace=True)
```

```python
flight.head()
```

```python
from sklearn.preprocessing import LabelEncoder
var=['Airline','Source','Destination','Additional_Info']
le=LabelEncoder()
for i in var:
  flight[i]=le.fit_transform(flight[i])
```

```python
flight.head()
```

```python
flight.dtypes
```

```python
flight['Total_Stops']=flight['Total_Stops'].astype(int)
```

```python
y=flight['Price']
x=flight.drop('Price',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)

```

```python
rf.score(x_train,y_train),rf.score(x_test,y_test)
```

```python
rf_pred = rf.predict(x_test)
```

```python
sns.distplot(y_test-rf_pred)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, rf_pred))
print('MSE:', metrics.mean_squared_error(y_test, rf_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, rf_pred)))
print('r2_score:', (metrics.r2_score(y_test, rf_pred)))
```
